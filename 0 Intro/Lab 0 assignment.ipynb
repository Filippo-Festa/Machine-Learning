```python
# Plot the data and the lines representing the output of the linregress and least square algorithms

plt.plot(data[:, 1], data[:, 3], 'o', label = 'year vs. price', markersize = 1)
plt.plot(data[:, 1], yearReg.intercept + yearReg.slope*data[:, 1], 'r', label='fitted line')
plt.legend()
plt.show()

```

```python
# linear regression with linregress (estimate price from power)    km_year_power_price

powerReg = stats.linregress(data[:, 2], data[:, 3])

plt.plot(data[:, 2], data[:, 3], 'o', label='power vs. price', markersize = 1)
plt.plot(data[:, 2], powerReg.intercept + powerReg.slope*data[:, 2], 'r', label='fitted line')
plt.legend()
plt.show()
```

```python
# linear regression with linregress (estimate price from km)

kilomReg = stats.linregress(data[:, 0], data[:, 3])

plt.plot(data[:, 0], data[:, 3], "o", label='kilometers vs. price', markersize = 1)
plt.plot(data[:, 0], kilomReg.intercept + kilomReg.slope*data[:, 0], 'r', label='fitted line')
plt.legend()
plt.show()
```

```python
# Have a look at the correlation coefficients to see which of the 3 features works better

print("Correlation Coefficient\nyear vs. price:", yearReg.rvalue, "\npower vs. price:", powerReg.rvalue ,"\nkilometers vs. price:", kilomReg.rvalue)
```
